```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
# Magnitude response for Normalize Lowpass Elliptic filter
for i in range(1, 7):
    b, a = sig.ellip(i, 1, 10, 1, analog=True, btype='low', output='ba')
    w, h = sig.freqs(b, a)
    print('n = {}'.format(i))
    print(b)
    print(a)
    plt.semilogx(w, np.power(abs(h), 2))
    plt.margins(0, 0.1)
    plt.grid(which='both', axis='both')
    plt.axhline(1, color='green')
    plt.title('Magnitude response for Normalize Elliptic filter for n={}'.format(i))
    plt.xlabel('w [rad/s]')
    plt.ylabel('|H(jw)|^2')
    plt.show()
```

```python
# Simple Example
wpass = 20
wstop = 30
n, ws = sig.ellipord(wpass, wstop, 3, 60, analog=True)
b, a = sig.ellip(n, 3, 60, ws, 'low', True)
w, h = sig.freqs(b, a)
print(n)
print(ws)
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.fill([0,wpass,wpass,0],[-3,-3,-99,-99],color='r',lw=0)
plt.fill([wstop,1000,1000,wstop],[-60,-60,100,100],color='r',lw=0)
plt.axis([1, 300, -100, 3])
plt.grid(which='both', axis='both')
plt.show()
```
